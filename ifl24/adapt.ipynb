```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
def unzip(xy):
    real, _, energy, threads = zip(*xy)
    real = list(map(lambda x: x / 1e9, real))
    energy = list(map(lambda y: y / 1e6, energy))
    print(f'Energy: {sum(energy):.2f}J, Runtime: {sum(real):.2f}s')
    return (real, energy, threads)
```

```python
def acc(xs):
    acc = 0
    for x in xs:
        acc += x
        yield acc
```

```python
def plot(title, data, bg_time, ymax=None):
    real, energy, t = unzip(data)

    ax = sns.lineplot(x=acc(real), y=energy)
    ax2 = ax.twinx()
    sns.lineplot(x=acc(real), y=t, color='r', alpha=0.7, ax=ax2)

    ax.set_title(title)
    ax.set_xlabel('Runtime (s)')
    ax.set_ylabel('Energy (J)')
    ax2.set_ylabel('#Threads')
    ax.set_ylim((0, ymax))
    ax2.set_ylim((0, 17))

    plt.axvline(x=60, color='g')
    plt.axvline(x=60 + bg_time / 1000, color='g')
    
    plt.show()
```
