```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('data/oracle.csv')
df
```

```python
total = df[df['oracle'] == False].copy()
total.reset_index(drop=True, inplace=True)

oracle_total = df[df['oracle'] == True]
oracle_total.reset_index(drop=True, inplace=True)

total['energy'] /= oracle_total['energy']
total['runtime'] /= oracle_total['runtime']
total['usertime'] /= oracle_total['usertime']
total
```

```python
print('Energy avg speedup:', total['energy'].sum() / len(total))
print('Runtime avg speedup:', total['runtime'].sum() / len(total))
print('Usertime avg speedup:', total['usertime'].sum() / len(total))
```
